```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
import math
```

```python
img= plt.imread('messi.jpg')#read as rgb
plt.figure(figsize=[15,10])

plt.subplot(1,3,1)

plt.imshow(img)

plt.title("RGB")

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
plt.title("BGR")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY),cmap='gray')
plt.title('GRAY')

plt.show()
```

```python
black_bg=np.zeros([2000,2000,3],dtype=np.uint8)
cv2.line(black_bg,(0,0),(2000,2000),(255,0,0),15)#red line

cv2.line(black_bg,(10,10),(500,700),(0,200,255),20)
cv2.line(black_bg,(500,700),(1000,500),(0,200,255),20)
cv2.line(black_bg,(1000,500),(700,150),(0,200,255),20)
cv2.line(black_bg,(700,150),(10,10),(0,200,255),20)

cv2.rectangle(black_bg,(1980,15),(1500,500),(0,255,0),20)
cv2.circle(black_bg,(1740,240),235,(0,0,255),-1)

font = cv2.FONT_HERSHEY_SIMPLEX 
org = (50, 50) 
fontScale = 1


center = (1000, 1000)
axes = (500, 250) 
angle = 140. 
cv2.ellipse(black_bg, center, axes, angle, 180., 360, (255,0,0),-1)

cv2.putText(black_bg,'OpenCV',(100,1900),cv2.FONT_HERSHEY_SIMPLEX,16,(255,255,255),thickness = 50
 ,bottomLeftOrigin=False)
plt.imshow(black_bg)
plt.show()
```
